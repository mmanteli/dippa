```python
import pandas
```

```python
df_en = pandas.read_csv("/scratch/project_2009498/dippa/zeroshot/results/en/test/en.tsv", sep=",")
df_fi = pandas.read_csv("/scratch/project_2009498/dippa/zeroshot/results/fi/test/fi.tsv", sep=",")
df_es = pandas.read_csv("/scratch/project_2009498/dippa/zeroshot/results/es/test/es.tsv", sep=",")
```

```python
df = pandas.concat([df_en, df_fi, df_es])

df.drop_duplicates(inplace=True)
df = df[df.Language!="Language"]
display(df)
print(df.columns)
df = df.rename(columns={" model":"Model"})
df = df.rename(columns={k:k.replace(" ", "") for k in df.columns.to_list()})
df.sort_values(by=["Language", "Model"], ascending=True, inplace=True)
display(df)


```

```python
df=df.applymap(lambda x : pandas.to_numeric(x,errors='ignore'))
df = df.round(decimals=3)
display(df)
```

```python
print(df.to_latex(float_format="%.3f"))
```

```python
import glob
```

```python
files = glob.glob('results/*/dev/*.txt')
print(files)
```

```python
data = {"Language":[], "Model":[], "Threshold":[]}
for f in files:
    with open(f, "r") as f_:
        z, th = f_.readline().split(": ")
        assert z == "best F1 th"
        th = float(th)
        langmod = f.split("/")[-1].replace(".txt", "")
        lang, mod = langmod.split("-", 1)
        data["Language"].append(lang)
        data["Model"].append(mod)
        data["Threshold"].append(th)
print(data)
```

```python
df2 = pandas.DataFrame.from_dict(data)
```

```python
df2.sort_values(by=["Language", "Model"], ascending=True, inplace=True)
```

```python
print(df2.to_latex(float_format="%.5g"))
```
